```python
import pandas as pd
import numpy as np
```

```python
ads = pd.read_table('ads_dataset.txt')
```

```python
col = ads.columns
ads.index = range(len(ads))
def getDfSummary(val):
    na_num=[]
    dst_num=[]
    for i in col:
        na_n=0
        na_n=sum(pd.isnull(ads[i].values))
        na_num.append(na_n)
        dst_n=len(ads.groupby(i).visit_freq)
        dst_num.append(dst_n)
    
    summary=ads.describe()
    summary=summary.T
    summary['NaN_count']=na_num
    summary['Distinct_Values']=dst_num
    summary = summary
    return summary
```
