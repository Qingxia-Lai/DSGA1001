```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,roc_auc_score
import random as rd
% matplotlib inline
```

```python
data = pd.read_csv("Cell2Cell_data.csv")
```

```python
#seperate the data: 80%: train; 20% test
rand = range(len(data))
div= int(len(data)*0.8)
rd.shuffle(rand)
train_index=rand[:div]
test_index=rand[div+1:]
train_set=data.iloc[train_index,:]
test_set=data.iloc[test_index,:]
```

```python
def TestAccuracy(TrainSet,TrainTarget,TestSet,TestTarget,split,leaf):
    clf = DecisionTreeClassifier(criterion='entropy',min_samples_split=split,min_samples_leaf=leaf)
    clf = clf.fit(TrainSet,TrainTarget)
    cm = confusion_matrix(clf.predict(TestSet),TestTarget)
    return (cm[0][0]+cm[1][1])/float(sum(sum(cm)))
```

```python
lab = 'churndep'
splits = np.arange(2,100,10)
leaves = np.arange(1,100,10)
```

```python
print splits,leaves
```

```python
res = []
for i in range(10):
    res.append([])

for i in range(10):
    for j in range(10):
        res[i].append(TestAccuracy(train_set.drop(lab,1),train_set[lab],test_set.drop(lab,1),test_set[lab],splits[i],leaves[j]))
```

```python
# strange 3D plot
from pylab import *
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1,1,1,projection = '3d')
p = ax.plot_surface(np.arange(10),np.arange(10),res,rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```

```python
#plot
fig = plt.figure(figsize = (13,10))
ax=fig.add_subplot(111)
plt.plot(splits,res[0],'b-',label='Min Samples Leaf = {}'.format(leaves[0]))
plt.plot(splits,res[1],'r-',label='Min Samples Leaf = {}'.format(leaves[1]))
plt.plot(splits,res[2],'c-',label='Min Samples Leaf = {}'.format(leaves[2]))
plt.plot(splits,res[3],'m-',label='Min Samples Leaf = {}'.format(leaves[3]))
plt.plot(splits,res[4],'y-',label='Min Samples Leaf = {}'.format(leaves[4]))
plt.plot(splits,res[5],'k-',label='Min Samples Leaf = {}'.format(leaves[5]))
plt.plot(splits,res[6],'b--',label='Min Samples Leaf = {}'.format(leaves[6]))
plt.plot(splits,res[7],'r--',label='Min Samples Leaf = {}'.format(leaves[7]))
plt.plot(splits,res[8],'c--',label='Min Samples Leaf = {}'.format(leaves[8]))
plt.plot(splits,res[9],'y--',label='Min Samples Leaf = {}'.format(leaves[9]))
plt.legend(loc=4)
ax.set_xlabel('Min Samples Split')
ax.set_ylabel('Test Set Accuracy ')
plt.title('Holdout Accuracy by Hyperparameters')
plt.show()
```
